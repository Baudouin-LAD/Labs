# Lab 32 : Encoding

```python
import pandas as pd

data_input = {
    'TransactionID': [1, 2, 3, 4, 5],
    'ClientID': [231, 765, 453, 231, 892],
    'Profession': ['Self-employed', 'students', 'Horeca', 'self-employed', 'finance'],
    'Bank_dep': ['009', '005', '007', '009', '003'],
    'Risk': ['high', 'high', 'medium', 'high', 'low'],
    'Number of credits': [2, 3, 5, 2, 3],
    'Revenue': [30200, 12700, 89400, 30200, 740000]
}

data = pd.DataFrame(data_input)
data
```

    Dummy Encoder


```python
pd.get_dummies(data)
```

    Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
le = LabelEncoder()

# List of columns to encode
cols_to_encode = ['Profession','Bank_dep']

# Encode the columns
for col in cols_to_encode:
    data[str(col)+'_label'] = le.fit_transform(data[col])+1

data
```

    Ordinal

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data['Risk'])
data['Risk_ordinal'] = le.transform(data['Risk']) + 1
data
```

```python
x = pd.DataFrame(data['Risk'])
x
from sklearn.preprocessing import OrdinalEncoder
ratings_order = ['low', 'medium', 'high']
oe = OrdinalEncoder(categories = [ratings_order], dtype=int)
oe.fit(x)
risk_ord = oe.transform(x)
risk_ord
data['risk_enc']=risk_ord  #new column with the values
data
```
