#### Question 1: Find the questions and answers of last month.

```python
from pprint import pprint
from datetime import datetime, timedelta
SITE = StackAPI("stackoverflow")
now = datetime.now()
last_month = now - timedelta(days=30)
from_date = int(last_month.timestamp())
questions = SITE.fetch("questions", sort="votes", fromdate=from_date)
answers = SITE.fetch("answers", sort="votes", fromdate=from_date)
pprint(questions)
pprint(answers)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
from pprint import pprint
from datetime import datetime, timedelta
today = datetime(2023, 1, 26, 0, 0, 0)
time = int(today.timestamp())
SITE.page_size = 10
SITE.max_pages = 1
questions = SITE.fetch("questions", fromdate=time, tagged="python", sort="votes", min=5)
pprint(questions)
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
badges = SITE.fetch('answers', ids=[6784, 6473])
pprint(badges)
```
