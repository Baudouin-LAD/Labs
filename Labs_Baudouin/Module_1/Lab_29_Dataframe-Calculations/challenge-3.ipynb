```python
customers = orders.groupby(['CustomerID'])['amount_spent'].sum().to_frame()
customers.head(20)
```

### Subproblem 2 & 3

```python
labels =['Normal','Regular','Prefered','VIP']
customers['Customer_label']=pd.qcut(customers['amount_spent'],4,labels=labels)
customers = customers.reset_index()
orders = orders.merge(customers[['CustomerID', 'Customer_label']], on='CustomerID', how='left')
orders.to_csv(r'/Users/Baudouin/Ironhack/Labs/Labs_Baudouin/Module_1/orders2.csv')
orders.head(20)
```

## Q2: How to identify which country has the most VIP Customers?

```python
customers_per_country = orders.pivot_table(index='Country',columns='Customer_label',values='CustomerID',aggfunc='count')
customers_per_country.sort_values(by=['VIP'])
```
