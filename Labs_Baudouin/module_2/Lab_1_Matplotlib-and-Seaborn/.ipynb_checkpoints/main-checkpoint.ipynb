#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
fig=plt.figure(figsize=(12,12))
plt.plot(x,y,color='#eb0000',label='x,y')
plt.plot(x,z,color='#00eb00',label='x,Z')
plt.title("phyton graph")
plt.xlabel('x variable')
plt.ylabel('y&z variables')
fig.legend(loc="lower right")
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y,color='#eb0000',label='x,y')
axes[1].plot(x,z,color='#00eb00',label='x,Z')
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, axes = plt.subplots(1,2,figsize=(9,10),)
axes[0].plot(x,y,color='#eb0000',label='x,y')
axes[0].set_title('x,y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
fig.suptitle('Subplots training', fontsize=16)
axes[1].plot(x,z,color='#00eb00',label='x,Z')
axes[1].set_title('x,z')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
fig.subplots_adjust(wspace=0.5)
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
a = x**2
b = np.exp(x)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.plot(x, a, color='#eb0000', label='x,x^2')
ax.legend(loc="upper right")
ax.set_yscale('linear')

ax2 = ax.twinx()
ax2.plot(x, b, color='#00eb00', label='x,e^x')
ax2.legend(loc="upper left")
ax2.set_yscale('log')

plt.title("scales")
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
a=fitbit.pivot_table(index='Months_encoded',values='Steps',aggfunc='median')
b=fitbit.pivot_table(index='Work_or_Weekend',values='Steps',aggfunc='median')
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].bar(b.index,b['Steps'],label='median daily steps weeks or weekend')
ax[0].legend(loc="upper right")
ax[0].set_xticks(ticks=[0,1],labels=["weekend","week"])
ax[1].bar(a.index,a['Steps'],label='median daily steps by month')
ax[1].legend(loc="upper right")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
b=fitbit.loc[:,('Steps','Minutes Lightly Active','Minutes Very Active','Minutes Sedentary')]
b#.iloc[:,1]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 3),sharey=True)

for i in range(len(b.columns.to_list())):
    if i!=0:
        ax[i-1].scatter(b.iloc[:,i],b.iloc[:,0],label=str(b.columns.to_list()[i]))
        ax[i-1].legend(loc='lower right',prop={'size': 6})
    
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.set(style="darkgrid")
# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
# assigning a graph to each ax
sns.boxplot(data=titanic,x="Age", ax=ax_box)
sns.histplot(titanic, x="Age", ax=ax_hist)
f.tight_layout()
# Remove x axis name for the boxplot
ax_box.set(xlabel='')
```
