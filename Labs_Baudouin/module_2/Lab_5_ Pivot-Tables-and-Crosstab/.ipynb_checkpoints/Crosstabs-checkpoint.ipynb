### 2. Use crosstabs to show the number of moovies per country

```python
movies_per_countries = data.pivot_table(index=['country'],values=['movie_title'],aggfunc='count')
```

```python
pd.set_option('display.max_rows', 65)
movies_per_countries
```

```python
pd.crosstab([data.country],aggfunc=[data.movie_title].valuecount())
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([data.director_name,data.genres], [data.country])
```
